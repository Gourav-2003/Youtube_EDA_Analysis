```python
pd.set_option("display.float_format", "{:.3f}".format)
category_subscriber_mean = df.groupby('category')['subscribers'].mean()
category_subscriber_mean
```

```python
category_subscriber_mean.sort_values(ascending = False , inplace = True )
```

```python
# top 7 plot 
category_subscriber_mean.head(7).plot(kind = 'bar')
plt.title("Mean subscriber by category")
plt.xlabel("category")
plt.ylabel("subscriber")
plt.show()
```

```python
# <!-- bottom 7 plot-->

category_subscriber_mean.tail(7).plot(kind = 'bar')
plt.title("Mean subscriber by category")
plt.xlabel("category")
plt.ylabel("subscriber")
plt.show()
```

```python
sns.boxenplot(data = df , x='channel_type' , y='highest_monthly_earnings' , palette='Set1')
plt.title("Distribution of highest_monthly_earnings by channel_type")
plt.xticks(rotation = 45)
plt.xlabel("channel_type")
plt.ylabel("highest_monthly_earnings")
plt.show()
```

```python
df.groupby(['category' , 'Country'])['subscribers'].sum().unstack().plot(kind='bar' , stacked = True)
plt.title("Subscribers Breakdown By Country and Categort")
plt.xlabel('Category')
plt.ylabel("Total subscribers")
plt.xticks(rotation = 90)
plt.legend(title = "Country", bbox_to_anchor = (1.05 ,1)  , loc='upper left')
plt.show()
```

```python
sns.scatterplot(data = df , x = 'Gross tertiary education enrollment (%)' , y='subscribers'  , palette = 'Set1')
plt.show()
```

```python
df.groupby('channel_type')['Unemployment rate'].mean().plot(kind = "bar")
plt.title("Avg unemployment rate By channel type")
plt.xlabel("channel type")
plt.ylabel("Unemployment rate")
plt.show()
```

```python
pd.set_option("display.float_format", "{:.0f}".format)


df.groupby(['created_year' ,'channel_type'])['subscribers'].mean().unstack().tail().plot(kind='bar' , width = 1)
plt.title("Avg subscribers By channel type and created year")
plt.xlabel("Created Year")
plt.ylabel("Average subscriber") 
plt.xticks(rotation = 45)
plt.legend(title="Channel Type" , bbox_to_anchor = (1.5 , 1) , loc = "upper right")
plt.show()

```

```python
pd.set_option("display.float_format", "{:.0f}".format)

df.groupby(['created_year' , 'created_month'])['subscribers'].sum().unstack().head().plot(kind = 'bar' , stacked = True)
plt.title("Total subscriber By created month and created year")
plt.xlabel("Created month")
plt.legend(title = "Created Year" , bbox_to_anchor = (1.3 , 1) , loc = "upper right")
plt.ylabel("Total subscribers")
plt.show()
```

```python
df.groupby(['created_year' , 'created_month'])['subscribers'].sum().unstack().head().plot(kind = 'area' , stacked = True)
plt.title("Total subscriber By created month and created year")
plt.xlabel("Created Year")
plt.legend(title = "Created Month" , bbox_to_anchor = (1.3 , 1) , loc = "upper right")
plt.ylabel("Total subscribers")
plt.show()
```

```python
df.groupby(['created_year' , 'Country'])['highest_yearly_earnings'].mean().unstack().head().plot(kind='line' , marker='o')
plt.title("Trend of Highest Yearly Earnings by Created Year and Country")
plt.xlabel("Created year")
plt.ylabel("Average Highest Yearly Earnings")
plt.legend(title = "Country", bbox_to_anchor = (1.05,1), loc = 'upper left')
plt.show()
```
